```python
# C7VWBi27oGc
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display_audio
import shutil

def show_mel(audio_path, figsize = (10, 4), xticks=10, max_duration = None):
    y, sr = librosa.load(audio_path)
    if max_duration != None:
        y = y[:sr*max_duration]
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=20)
    S_db = librosa.power_to_db(S, ref=np.max)

    plt.figure(figsize=figsize)
    librosa.display.specshow(S_db, x_axis='time', y_axis='mel', fmax=8000)
    plt.xticks(np.arange(0, y.shape[0]/sr, xticks))

    # plt.colorbar(format='%+2.0f dB')
    plt.show()


# -8S9TKWIOSc  1wIIDGeNcXI 1EHKoWs3NFk 31GOxPXDNkk
# 94mMR_HhFac C7VWBi27oGc

audio_name = '7SGOZRgBUDs'
fc_audio_path = f'/home/chengxin/chengxin/unav100/generated_audios/foleycrafter/10/{audio_name}.wav'
fc_audio_path_ = f'/home/chengxin/chengxin/unav100/generated_audios/foleycrafter_/dur10/{audio_name}.wav'
# sd_audio_path = f'/home/chengxin/chengxin/unav100/generated_audios/stablev2a/model14_ss30_cfg5/{audio_name}.wav'
sd_audio_path_ = f'/home/chengxin/chengxin/unav100/generated_audios/stablev2a_/model8_ss30_cfg5/{audio_name}.wav'
df_audio_path_ = f'/home/chengxin/chengxin/unav100/generated_audios/difffoley/10/{audio_name}.wav'
tv_audio_path_ = f'/home/chengxin/chengxin/unav100/generated_audios/tiva/10/{audio_name}.wav'
gt_audio_path = f'/home/chengxin/chengxin/unav100/generated_audios/test/10/{audio_name}.wav'

shutil.copy(f'/home/chengxin/chengxin/unav100/dataset/test/{audio_name}.mp4', './asset/test.mp4')
# shutil.copy(f'/home/chengxin/chengxin/unav100/dataset/test/{audio_name}.mp4', f'/home/chengxin/chengxin/FoleyCrafter/examples/unav/{audio_name}.mp4')

show_mel(gt_audio_path)
show_mel(sd_audio_path_)
# display_audio.Audio(gt_audio_path)
```

```python
import os
import subprocess
import random
import shutil
from tqdm import tqdm

def replace_audio(input_video, input_audio, output_video):
    ffmpeg_cmd = [
        'ffmpeg',
        '-i', input_video,        # 
        '-i', input_audio,        # 
        '-c:v', 'copy',           # 
        '-c:a', 'aac',            # aac
        '-map', '0:v:0',          # 
        '-map', '1:a:0',          # 
        '-strict', 'experimental', # 
        '-loglevel', 'error',
        output_video,              # 
        '-y'
    ]

    try:
        subprocess.run(ffmpeg_cmd, check=True)
    except subprocess.CalledProcessError as e:
        print(f": {e}")








# sd19_dir = '/home/chengxin/chengxin/unav100/generated_audios/stablev2a_/model19_ss60_cfg5'
# sd8_dir = '/home/chengxin/chengxin/unav100/generated_audios/stablev2a_/model8_ss60_cfg5'
# target_dir = '/home/chengxin/chengxin/stable-v2a/demo/humantest'
# video_dir = '/home/chengxin/chengxin/unav100/dataset/test'

# spec_dir = '/home/chengxin/chengxin/unav100/generated_audios/specvqgan/10'
# im2wav_dir = '/home/chengxin/chengxin/unav100/generated_audios/im2wav/10'
# foley_dir = '/home/chengxin/chengxin/unav100/generated_audios/foleycrafter_/dur10'
# df_dir = '/home/chengxin/chengxin/unav100/generated_audios/difffoley/10'
# tv_dir = '/home/chengxin/chengxin/unav100/generated_audios/tiva/10'




sd19_dir = '/home/chengxin/chengxin/VGGSound/generated_audios/stablev2a_/model19_cfg5'
sd8_dir = '/home/chengxin/chengxin/VGGSound/generated_audios/stablev2a_/model8_cfg5'
target_dir = '/home/chengxin/chengxin/stable-v2a/demo/humantest'
video_dir = '/home/chengxin/chengxin/VGGSound/dataset/test/10'

spec_dir = '/home/chengxin/chengxin/VGGSound/generated_audios/specvqgan/10'
im2wav_dir = '/home/chengxin/chengxin/VGGSound/generated_audios/im2wav/10'
foley_dir = '/home/chengxin/chengxin/VGGSound/generated_audios/foleycrafter/10'
df_dir = '/home/chengxin/chengxin/VGGSound/generated_audios/difffoley/10'
tv_dir = '/home/chengxin/chengxin/VGGSound/generated_audios/tiva/10'


file_names = os.listdir(sd19_dir)
random.shuffle(file_names)




for file_name in tqdm(file_names[:50]):
    video_name = file_name.replace(".wav", ".mp4")
    video_path = os.path.join(video_dir, video_name)


    audio_path = os.path.join(sd19_dir, file_name)
    output_path = os.path.join(target_dir, video_name.replace(".mp4", "_sd19.mp4"))
    replace_audio(video_path, audio_path, output_path)


    audio_path = os.path.join(sd19_dir, file_name)
    output_path = os.path.join(target_dir, video_name.replace(".mp4", "_sd8.mp4"))
    replace_audio(video_path, audio_path, output_path)

    audio_path = os.path.join(im2wav_dir, file_name)
    output_path = os.path.join(target_dir, video_name.replace(".mp4", "_im.mp4"))
    replace_audio(video_path, audio_path, output_path)


    audio_path = os.path.join(foley_dir, file_name)
    output_path = os.path.join(target_dir, video_name.replace(".mp4", "_fc.mp4"))
    replace_audio(video_path, audio_path, output_path)


    audio_path = os.path.join(df_dir, file_name)
    output_path = os.path.join(target_dir, video_name.replace(".mp4", "_df.mp4"))
    replace_audio(video_path, audio_path, output_path)


    audio_path = os.path.join(tv_dir, file_name)
    output_path = os.path.join(target_dir, video_name.replace(".mp4", "_tv.mp4"))
    replace_audio(video_path, audio_path, output_path)

    audio_path = os.path.join(spec_dir, file_name)
    output_path = os.path.join(target_dir, video_name.replace(".mp4", "_spc.mp4"))
    replace_audio(video_path, audio_path, output_path)

    video_path = os.path.join(video_dir, file_name.replace(".wav", ".mp4"))
    gt_output_path = os.path.join(target_dir, file_name.replace(".wav", f"_gt.mp4"))
    shutil.copy(video_path, gt_output_path)


    
    
```

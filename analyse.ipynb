```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display_audio
import shutil

def show_mel(audio_path, figsize = (10, 4), xticks=10, max_duration = None):
    y, sr = librosa.load(audio_path)
    if max_duration != None:
        y = y[:sr*max_duration]
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=20)
    S_db = librosa.power_to_db(S, ref=np.max)

    plt.figure(figsize=figsize)
    librosa.display.specshow(S_db, x_axis='time', y_axis='mel', fmax=8000)
    plt.xticks(np.arange(0, y.shape[0]/sr, xticks))

    # plt.colorbar(format='%+2.0f dB')
    plt.show()


# -8S9TKWIOSc  -9HoMjD6Opk  1wIIDGeNcXI 1EHKoWs3NFk 31GOxPXDNkk
# 94mMR_HhFac C7VWBi27oGc

audio_name = 'C7VWBi27oGc'
fc_audio_path = f'/home/chengxin/chengxin/unav100/generated_audios/foleycrafter/10/{audio_name}.wav'
fc_audio_path_ = f'/home/chengxin/chengxin/unav100/generated_audios/foleycrafter_/dur10/{audio_name}.wav'
# sd_audio_path = f'/home/chengxin/chengxin/unav100/generated_audios/stablev2a/model14_ss30_cfg5/{audio_name}.wav'
sd_audio_path_ = f'/home/chengxin/chengxin/unav100/generated_audios/stablev2a_/model8_ss30_cfg5/{audio_name}.wav'
df_audio_path_ = f'/home/chengxin/chengxin/unav100/generated_audios/difffoley/10/{audio_name}.wav'
tv_audio_path_ = f'/home/chengxin/chengxin/unav100/generated_audios/tiva/10/{audio_name}.wav'
gt_audio_path = f'/home/chengxin/chengxin/unav100/generated_audios/test/10/{audio_name}.wav'

shutil.copy(f'/home/chengxin/chengxin/unav100/dataset/test/{audio_name}.mp4', './asset/test.mp4')
# shutil.copy(f'/home/chengxin/chengxin/unav100/dataset/test/{audio_name}.mp4', f'/home/chengxin/chengxin/FoleyCrafter/examples/unav/{audio_name}.mp4')

show_mel(gt_audio_path)
show_mel(sd_audio_path_)
# display_audio.Audio(gt_audio_path)
```

```python
import json
import torch
from stable_audio_tools import create_model_from_config, replace_audio, save_audio
from stable_audio_tools.data.dataset import VideoFeatDataset, collation_fn
from stable_audio_tools.training.training_wrapper import DiffusionCondTrainingWrapper
from stable_audio_tools.inference.generation import generate_diffusion_cond, generate_diffusion_cond_from_path
from moviepy.editor import VideoFileClip
from safetensors.torch import load_file


model_config_file = './stable_audio_tools/configs/model_config.json'
# config_t2a: t2a      config:v2a     config1:new v2a
with open(model_config_file) as f:
    model_config = json.load(f)

sample_rate = model_config["sample_rate"]
sample_size = model_config["sample_size"]
fps = model_config["fps"]
sample_size = None

model = create_model_from_config(model_config)
model_dir = './weight/StableAudio/2024-08-04 02:52:24'
model_name = 'epoch=27-step=2818'  
state_dict = load_file(f'{model_dir}/{model_name}.safetensors')
print(f'{model_dir}/{model_name}.safetensors')
# state_dict = load_file(f'./weight/StableAudio/model.safetensors')
model.load_state_dict(state_dict, strict=False)

sample_rate, fps, sample_size
```

```python

device = 0
conditioning = {
    # 'seconds_start': [0, 0],
    # 'seconds_total': [10, 10],
    'feature': [f'/home/chengxin/chengxin/unav100/dataset/test/{audio_name}.mp4']
}
video_clip = VideoFileClip(conditioning['feature'][0])
seconds_total = int(video_clip.duration)
seconds_total = 10

output = generate_diffusion_cond(
    model = model.to(device),
    steps=150,
    cfg_scale=7,
    conditioning=conditioning,
    sample_size=int(sample_rate*seconds_total),
    batch_size=len(conditioning['feature']),
    sigma_min=0.3,
    sigma_max=500,
    sampler_type="dpmpp-3m-sde",
    device=device
)

audio_path = "./asset/test.wav"
waveform = output[0:1,...,:int(seconds_total*sample_rate)]
# print(output.shape, output[idx:idx+1].shape, waveform.shape)
save_audio(waveform, audio_path, sample_rate)

show_mel(audio_path,figsize=(3,4))
# display_audio.Audio(audio_path)
```

```python
import matplotlib.pyplot as plt


bin_size = 10  
max_value = int(max(audio_duration))+1  
bins = [i for i in range(0, max_value + bin_size, bin_size)]  

counts = [0] * len(bins)  
for duration in audio_duration:
    for i in range(len(bins) - 1):
        if bins[i] <= duration < bins[i + 1]:
            counts[i] += 1
            break

plt.bar(range(len(counts)), counts)
plt.title('Number Distribution by Interval')
plt.xlabel('Interval')
plt.ylabel('Count')
plt.xticks(range(len(counts)), [f'{bins[i]}-{bins[i+1]}' if i < len(bins) - 1 else f'{bins[-1]}+' for i in range(len(bins))])
plt.show()
```
